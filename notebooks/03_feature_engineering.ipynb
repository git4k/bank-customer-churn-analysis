# Step 3: Feature Engineering
Create new features and encode categorical variables

```python
import pandas as pd
import numpy as np
```

```python
# Load cleaned data
df = pd.read_csv('../data/processed/bank_churn_clean.csv')
print(f"Initial shape: {df.shape}")
print(f"Initial columns: {list(df.columns)}")
```

```python
# Create RFM-style features
df['BalanceSalaryRatio'] = df['Balance'] / (df['EstimatedSalary'] + 1)
df['TenureAgeRatio'] = df['Tenure'] / (df['Age'] + 1)
df['ProductsPerTenure'] = df['NumOfProducts'] / (df['Tenure'] + 1)

print("\nNew features created:")
print(df[['BalanceSalaryRatio', 'TenureAgeRatio', 'ProductsPerTenure']].head())
```

```python
# Encode categorical variables
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True)

print(f"\nFinal shape: {df.shape}")
print(f"Final columns: {list(df.columns)}")
```

```python
# Check for any infinite or NaN values
print("\nInfinite values:")
print(np.isinf(df.select_dtypes(include=[np.number])).sum())

print("\nNaN values:")
print(df.isnull().sum())

# Replace infinite values with NaN and then fill
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)
```

```python
# Save engineered features
df.to_csv('../data/processed/bank_churn_features.csv', index=False)
print("\nFeature-engineered data saved to data/processed/bank_churn_features.csv")
df.head()
```
